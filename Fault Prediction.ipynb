```python
#Import Modules and DataSet
import pandas as pd
import numpy as np

df = pd.read_csv('E:/DataSets/Pump Sensor Data/sensor.csv')
df.head()
```

```python
#Delete the Count(1st) Column
del df['Unnamed: 0']
# convert time into index 
df['index'] = pd.to_datetime(df['timestamp'])
df.index = df['index']
# delete the colunmns 
del df['index']
del df['timestamp']
df.head(2)
```

```python
df['sensor_15'].nunique() # no unique - complete zero
# drop the column 
df.drop(['sensor_15'], axis=1, inplace = True)
print(df.shape)
df.info()
```

```python
# machine status - no null 
# we will drop na in whole dataframe 
df['sensor_00'].isna().sum()
```

```python
df['machine_status'].unique()#'NORMAL', 'BROKEN', 'RECOVERING' 
df['machine_status'].value_counts()
```

```python
# apply label encoder to encode the machine status
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['machine_status'] = le.fit_transform(df['machine_status'])
df['machine_status'].value_counts()

# 1 - normal 
# 2 - recovering 
# 0 - broken
```

```python
#  look on complete data frame when device is broken
df_broken = df[df.machine_status ==0]
df_broken
```

```python
#used ffill(Forward Fill) method to fill the missing values
#Since the Data is a timeseries data, we use ffill
df = df.fillna(method='ffill')
```

```python
#Building The Classifier
X = df.drop(['machine_status'], axis=1)
print("Data Shape ",X.shape)
Y = df['machine_status']
print("Class Shape = ",Y.shape)
```

```python
# apply the logistic regression 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.30, random_state = 42)
model = LogisticRegression(max_iter = 400).fit(X_train, y_train)
```

```python
# predict
y_pred = model.predict(X_test)
# evaluate the model
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
cm = pd.crosstab(y_test,y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
cm
```

```python
# accuracy is not a good metrics for Anomaly detection and imblaanced dataset
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
# Classification Report
cr = classification_report(y_pred, y_test)
print(cr)
```

```python
#Even though the Accuracy Score for the model is really high(99.67%), the precision and recall for the minority class is really
#low. Hence it is not a good model. So we Apply SMOTE algorithm to balance the data and then Check for the values
!pip install imblearn
```

```python
from imblearn.over_sampling import SMOTE 
sm = SMOTE(random_state = 12,k_neighbors =4) 
X_train_res, y_train_res = sm.fit_sample(X_train, y_train)
print('After OverSampling, the shape of train_X: {}'.format(X_train_res.shape)) 
print('After OverSampling, the shape of train_y: {} \n'.format(y_train_res.shape))

print("After OverSampling, counts of label '0': {}".format(sum(y_train_res == 0))) 
print("After OverSampling, counts of label '1': {}".format(sum(y_train_res == 1))) 
print("After OverSampling, counts of label '2': {}".format(sum(y_train_res == 2))) 
```

```python
model1=LogisticRegression(max_iter=400).fit(X_train_res,y_train_res)
```

```python
# predict
y_pred = model1.predict(X_test)
# evaluate the model
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
cm = pd.crosstab(y_test,y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
cm
```

```python
# accuracy is not a good metrics for Anomaly detection and imblaanced dataset
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
# Classification Report
cr = classification_report(y_pred, y_test)
print(cr)
```

```python
#Even though the Accuracy has reduced a bit(99.67% to 99.24%)the precision and recall for the minority class has increased
#significantly. Hence the oversampled Model is better than the Previous model
```
